## Hedef Kitle
- 9 eşsiz değeri var
- Veriler cinsiyet ve yaş aralığı değerlerinden meydana geliyor
- All ages şeklinde doğrudan cinsiyet veya yaş belirtmeyen bir değer mevcut

### Yapılacak İşlemler:
- Veriler Cinsiyet ve Yaş değerlerine ayrılarak hedef cinsiyet ve hedef yaş değişkenleri oluşturulacak
- All Ages değerinin, bütün yaş ve cinsiyetleri ifade ettiği varsayılacak
- Gözlem değerleri şu şekilde güncellenecek:
  1. Women: Kadın
  2. Men: Erkek
  3. All Ages: Hepsi 18-60
- Hedef kitle değişkeni silinecek
- Hedef Cinsiyet ve Hedef Yaş değişkenleri kategori türüne dönüştürülecek

```python
# hedef kitle eşsiz değerleri
data["hedef_kitle"].unique()
```

```python
# yaş ve cinsiyet değerlerini ayırma
data[["hedef_cinsiyet", "hedef_yas"]] = data["hedef_kitle"].str.split(" ", expand=True)
```

```python
# gözlem değerleri güncellemesi
eski_deger = ["Women", "Men", "All"]
yeni_deger = ["Kadın", "Erkek", "Hepsi"]
for eski, yeni in zip(eski_deger, yeni_deger):
    data["hedef_cinsiyet"] = data["hedef_cinsiyet"].str.replace(eski, yeni)
data["hedef_yas"] = data["hedef_yas"].str.replace("Ages", "18-60")
```

```python
# hedef cinsiyet ve hedef yaş değişkenlerini kategori türüne dönüştürme
data["hedef_cinsiyet"] = data["hedef_cinsiyet"].astype("category")
data["hedef_yas"] = data["hedef_yas"].astype("category")
```

```python
# hedef kitle değişkenini silme
data.drop("hedef_kitle", axis = 1, inplace=True)
```

```python
# değişen değerlerin kontrolü
data["hedef_cinsiyet"].unique(), data["hedef_yas"].unique()
```

## Eksik Veri Oluşturma Fonksiyonu


```python
def add_random_missing_values(dataframe: pd.DataFrame,
                              missing_rate: float = 0.05) -> pd.DataFrame:
    """Turns random values to NaN in a DataFrame.

    To use this function, you need to import pandas, numpy and random libraries.

    Args:
        dataframe (pd.DataFrame): DataFrame to be processed.
        missing_rate (float): Percentage of missing value rate in float format. Defaults 0.05

    Returns:
        df_missing (pd.DataFrame): Processed DataFrame object.

    """
    # Get copy of dataframe
    df_missing = dataframe.copy()

    # Obtain size of dataframe and number total number of missing values
    df_size = dataframe.size
    num_missing = int(df_size * missing_rate)

    # Get random row and column indexes to turn them NaN
    for _ in range(num_missing):
        row_idx = random.randint(0, dataframe.shape[0] - 1)
        col_idx = random.randint(0, dataframe.shape[1] - 1)

        df_missing.iat[row_idx, col_idx] = np.nan

    return df_missing
```

```python
data = add_random_missing_values(dataframe= data, missing_rate= 0.02)
```

```python
# eksik gözlemlerin sayısını bulalım
data.isnull().sum()
```

```python
# kategorik değişkenlerin eksik gözlem değerlerini mod ile dolduralım
kategorik_sutunlar = ["kampanya_amaci", "kampanya_suresi", "kullanilan_kanal", "konum", "dil", "musteri_segmenti", "sirket", "hedef_cinsiyet", "hedef_yas", "gun_ismi", "ay", "gun"]
for i in kategorik_sutunlar:
    mode_value = data[i].mode()[0]
    data[i].fillna(mode_value, inplace=True)

# numerik değişkenlerin eksik gözlem değerlerini ortalama ile dolduralım
numerik_sutunlar = ["donusum_orani", "edinim_maliyeti", "yatirim_getirisi", "tiklama_sayisi", "gosterim_sayisi", "katilim_puani"]
for i in numerik_sutunlar:
    mean_value = data[i].mean()
    data[i].fillna(mean_value, inplace=True)
```

# İstatistiksel Testler ve Yorumlama

```python
from scipy import stats
```

```python
# Shapiro-Wilk testi - Normallik Testi (3-5000 arasında gözlem sayısına sahip veri setleri için uygun)
# büyük bir veri seti ile çalıştığım için 5000 rastgele örnek seçerek, edinim_maliyeti değişkeninin normal dağılıp dağılmadığını test ettim
# veriler normal  dağılıma uymadığından buna uygun olarak, kruskal wallis istatistiği yapılacaktır
orneklem = data['edinim_maliyeti'].sample(5000)

statistic, p_value = stats.shapiro(orneklem)

print(f'Shapiro-Wilk İstatistiği: {statistic}, p-değeri: {p_value}')

# Sonuçların yorumlanması
alpha = 0.05
if p_value < alpha:
    print("H0 hipotezini reddediyoruz: Veri normal dağılıma uymuyor.")
else:
    print("H0 hipotezini reddedemiyoruz: Veri normal dağılıma uyuyor.")
```

```python
# kruskal test istatistiği
stat, p_value = stats.kruskal(
    data[data['kampanya_suresi'] == 15]['edinim_maliyeti'],
    data[data['kampanya_suresi'] == 30]['edinim_maliyeti'],
    data[data['kampanya_suresi'] == 45]['edinim_maliyeti'],
    data[data['kampanya_suresi'] == 60]['edinim_maliyeti']
)

print(f'Kruskal-Wallis İstatistiği: {stat}, p-değeri: {p_value}')

# Sonuçların yorumlanması
alpha = 0.05
if p_value < alpha:
    print("H0 hipotezini reddediyoruz: Kampanya süresinin edinim maliyeti üzerinde anlamlı bir etkisi var.")
else:
    print("H0 hipotezini reddedemiyoruz: Kampanya süresinin edinim maliyeti üzerinde anlamlı bir etkisi yok.")
```

**Yorumlama**
1. Kruskal-Wallis İstatistiği
Yüksek Değer: Kruskal-Wallis istatistiği, gruplar arasındaki sıralı farkları yansıtır. Burada elde edilen 256766.77 değeri, gruplar arasında belirgin bir fark olduğunu gösterir.
2. P-değeri
Anlamlılık: P-değeri 0.0, istatistiksel olarak son derece anlamlı bir sonuçtur. Bu, elde edilen sonucun rastgele oluşma olasılığının son derece düşük olduğunu gösterir.
Alpha Düzeyi: Alpha (α) değeri 0.05 olarak belirlenmiş. P-değerinin 0.0 olması, bu değerin çok altında olduğu için, H0 hipotezini reddetmemiz gerektiği anlamına gelir. Bu, kampanya süresinin edinim maliyeti üzerinde istatistiksel olarak anlamlı bir etkisi olduğu anlamına gelir.

Bu sonuçlar, farklı kampanya sürelerine sahip grupların edinim maliyetleri arasında anlamlı farklılıklar olduğunu ortaya koymaktadır. Yani, kampanya süresi arttıkça edinim maliyetinin de değiştiğini söylenir. Bu durum, kampanya yönetimi ve optimizasyonu açısından önemli bilgiler sunar. 

```python
# spearman korelasyon katsayısı ile ilişkinin hakkında daha ayrıntılı bilgi alınır
corr, p_value = stats.spearmanr(data['kampanya_suresi'], data['edinim_maliyeti'])

print(f"Spearman Korelasyon Katsayısı: {corr:.2f}")
print(f"P-değeri: {p_value:.4f}")
```

**Yorumlama** 
1. Spearman Korelasyon Katsayısı (0.93)
Değerin Yüksekliği: 0.93, iki değişken arasında çok güçlü bir pozitif ilişki olduğunu gösterir. Bu, kampanya süresi ile edinim maliyeti arasında sıralı bir ilişki olduğu anlamına gelir. Yani, kampanya süresi arttıkça edinim maliyetinin de arttığını söyleyebiliriz.
Sıralı İlişki: Spearman korelasyonu, verinin sıralı ilişkisini ifade eder. Bu durumda, yüksek kampanya sürelerine sahip olan gözlemler genellikle daha yüksek edinim maliyetlerine sahiptir.
2. P-değeri (0.0000)
Anlamlılık: P-değeri, elde edilen korelasyonun istatistiksel olarak anlamlı olup olmadığını değerlendirir. Genellikle, 0.05'ten küçük bir p-değeri, ilişki için istatistiksel olarak anlamlı kabul edilir.
Sonuç: P-değerinin 0.0000 olması, kampanya süresi ile edinim maliyeti arasındaki ilişkinin çok güçlü ve anlamlı olduğunu gösterir. Yani, bu ilişkinin rastgele oluşma olasılığı son derece düşüktür.
